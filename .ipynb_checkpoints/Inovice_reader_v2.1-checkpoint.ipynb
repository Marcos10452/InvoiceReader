```python
class MainGUI:
    def __init__(self, master):
        self.master = master
        self.master.title("Invoice editor")
        # Adjust size
        self.HEIGHT=800
        self.WIDTH=1200
        self.master.geometry(str(self.WIDTH)+"x"+str(self.HEIGHT))
   
        #------Data Frame---------
        self.fr_data = tk.Frame(self.master,relief=tk.FLAT, bd=2,width=int(self.WIDTH),height=int(self.HEIGHT*0.21), bg='grey')
        self.fr_data.grid(row=0, column=0,sticky="w")
        self.fr_data.grid_propagate(0)

        #------Image Frame---------
        #Color #rgb #rrggbb #rrrgggbbb
        self.fr_img = tk.Frame(self.master,relief=tk.FLAT, bd=2,width=self.WIDTH,height=self.HEIGHT*0.79, bg='lightgrey')
        self.fr_img.grid(row=1, column=0)
        self.fr_img.grid_propagate(False)
      

        self.new_path=""
        self.filename=""
        self.new_prov_code=6
        self.new_prov_num=0
        self.auxpage=[0]
        self.name=""
        

        self.dffinal= pd.DataFrame()
        self.dfresult=pd.DataFrame()
        self.df=pd.DataFrame()

        self.menu_bar()
        self.label_place()
        self.fillin_place()
       
        self.pic_place("")
        
    def menu_bar(self):  
        #Menu
        menubar = tk.Menu(self.master, relief=tk.RAISED, bd=2)
        filemenu = tk.Menu(menubar, tearoff = 0)
        filemenu.add_command(label = "Open", command = self.open_file)
    
        filemenu.add_separator()
        filemenu.add_command(label = "Exit", command = self.master.destroy)
        menubar.add_cascade(label = "File", menu = filemenu)


        self.master.config(menu = menubar)

  

    def label_place(self):  
        #Label
        #The height of the widget in lines (not pixels!), measured according to the current font size.
        #The width of the widget in characters (not pixels!), measured according to the current font size.
        lb_provcode=tk.Label(self.fr_data, text="Provider's code" ,relief = tk.RAISED)
        lb_provcode.grid(row=0, column=0, sticky="ew", padx=5, pady=5)
        lb_invnum=tk.Label(self.fr_data, text='Invoice Number',relief = tk.RAISED)
        lb_invnum.grid(row=1, column=0, sticky="ew", padx=5, pady=5)
        lb_invdate=tk.Label(self.fr_data,text='Invoice Date',relief = tk.RAISED)
        lb_invdate.grid(row=2, column=0, sticky="ew", padx=5, pady=5)
        lb_caenum=tk.Label(self.fr_data, text='CAE Number',relief = tk.RAISED)
        lb_caenum.grid(row=3, column=0, sticky="ew", padx=5, pady=5)
        lb_caedate=tk.Label(self.fr_data, text='CAE Date',relief = tk.RAISED)
        lb_caedate.grid(row=4, column=0, sticky="ew", padx=5, pady=5)
        lb_tlamount=tk.Label(self.fr_data, text='Total Amount',relief = tk.RAISED)
        lb_tlamount.grid(row=5, column=0, sticky="ew", padx=5, pady=5)
        #self.fillin_place(self.fr_data)

    def fillin_place(self): 
        #The height of the widget in lines (not pixels!), measured according to the current font size.
        #The width of the widget in characters (not pixels!), measured according to the current font size. 
        self.ms_provcode = tk.Entry( self.fr_data, text = "", relief = tk.SUNKEN,width=30 )
        self.ms_provcode.grid(row=0, column=1, sticky="ew", padx=5, pady=5)
        self.ms_invnum = tk.Entry( self.fr_data, text = "", relief = tk.SUNKEN,width=30 )
        self.ms_invnum.grid(row=1, column=1, sticky="ew", padx=5, pady=5)
        self.ms_invdate = tk.Entry( self.fr_data, text = "", relief = tk.SUNKEN,width=30 )
        self.ms_invdate.grid(row=2, column=1, sticky="ew", padx=5, pady=5)
        self.ms_caenum = tk.Entry( self.fr_data, text= "", relief = tk.SUNKEN,width=30 )
        self.ms_caenum.grid(row=3, column=1, sticky="ew", padx=5, pady=5)
        self.ms_caedate = tk.Entry( self.fr_data, text = "", relief = tk.SUNKEN,width=30 )
        self.ms_caedate.grid(row=4, column=1, sticky="ew", padx=5, pady=5)
        self.ms_tlamount = tk.Entry( self.fr_data, text = "", relief = tk.SUNKEN,width=30 )
        self.ms_tlamount.grid(row=5, column=1, sticky="ew", padx=5, pady=5)

      
    def pic_place(self,img):  
        image_window = ScrollableImage(self.fr_img, image=img, scrollbarwidth=20, 
                               width=int(self.WIDTH*0.98), height=int(self.HEIGHT*0.75))
        image_window.grid(row=0, column=0, sticky="ew")
    
    def open_file(self):
        """Open a file for editing."""
        global PdfFile
        
        PdfFile=FileManager(PATHINVOICE,providerName)
        
        self.new_path=PdfFile.look_path()
        self.new_prov_num, self.new_prov_code,self.name= PdfFile.get_provider_data(self.new_path)
      
        list_path=[]
        list_path.append(self.new_path)
        self.df,_,_,_= PdfFile.open_file(list_path)
        self.filename=self.new_path.replace(PATHINVOICE,"")

        img = Image.open(self.new_path+".png")
        img = ImageTk.PhotoImage(img)

        self.pic_place(img)
        self.invoice_items()

    def invoice_items(self):  


        ###________________ Data FRAME processing data____________________
        self.df=process_df(self.df,self.new_prov_code)
        ###___________________________   END  _____________________________
        ###________________ Getting invoice prediction_____________________
        self.df['Class']=prediction_func(self.df)
        ###___________________________   END  _____________________________
        
        #Create array with 'x0','top','Providercode','page_number'&'text' columns 
        #REVISAR, AL PARECER NO TIENE SENTIDO
        testarray=np.array(self.df[['x0','top','Providercode','page_number','text','Class']].values)

        ### First table- Invoice number and CAE
        lpaginas=[0]
        self.auxpage=[0]
        lnumerofact=[""]
        lfechafact=[""]
        lnumerocae=[""]
        lfechacae=[""]
        ltotal=[""]

        #testarray should be sorted by pagenumber before in order to get consecutive pagenumbers
        for forarray in (testarray):
            lnumerofact=check_option2(forarray[5],1,lnumerofact,forarray[4])#Invoice Nº always first page
            lfechafact=check_option2(forarray[5],2,lfechafact,forarray[4])#Invoice date always first page
            lnumerocae=check_option2(forarray[5],3,lnumerocae,forarray[4])
            lfechacae=check_option2(forarray[5],4,lfechacae,forarray[4])
            ltotal=check_option2(forarray[5],5,ltotal,forarray[4])  
            lpaginas.append(forarray[3])

        self.auxpage[0]=max(lpaginas)

        #reset entry in GUI
        self.ms_provcode.delete(0,tk.END)
        self.ms_provcode.delete(0,tk.END)
        self.ms_invnum.delete(0,tk.END)
        self.ms_invdate.delete(0,tk.END)
        self.ms_caenum.delete(0,tk.END)
        self.ms_caedate.delete(0,tk.END)
        self.ms_tlamount.delete(0,tk.END)
        
        #LOAD entry with data
        self.ms_provcode.insert(-1,self.new_prov_num)
        self.ms_invnum.insert(-1,RegExpression(lnumerofact[0],list_providers[self.name][0]))
        self.ms_invdate.insert(-1,RegExpression(lfechafact[0],list_providers[self.name][1]))
        self.ms_caenum.insert(-1,RegExpression(lnumerocae[0],list_providers[self.name][2]))
        self.ms_caedate.insert(-1,RegExpression(lfechacae[0],list_providers[self.name][3]))
        self.ms_tlamount.insert(-1,ltotal)
    

```

```python

def main(): 

    window = tk.Tk()
    app = MainGUI(window)
    window.mainloop()

if __name__ == '__main__':
    main()
```
